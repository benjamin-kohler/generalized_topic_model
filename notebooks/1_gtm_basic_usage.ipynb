```python
import pandas as pd
df = pd.read_csv('../data/us_congressional_record/us_congress_speeches_sample_1000.csv')
df.head()
```

```python
import sys
sys.path.append('../gtm/')
from corpus import GTMCorpus
from gtm import GTM

train_dataset = GTMCorpus(
    df, 
    prevalence = "~ Rep", # party affiliation affects the probability of talking about a topic
    content = "~ Rep", # party affiliation affects how topics are discussed
    vectorizer_args={
        'ngram_range':(1, 1), # focus on unigrams
        'max_df':0.99, 
        'min_df':0.001
    }
)
```

```python
# Train the model
tm = GTM(
    train_dataset, 
    n_topics=20,
    update_prior=True,
    dropout=0.2,
    w_prior=None
)
```

```python
print(
    "\n".join(
        [
            "{}: {}".format(str(k), str(v))
            for k, v in tm.get_topic_words(topK=5).items()
        ]
    )
)
```
